### Some  shortcuts
1. Ctrl+Enter: Run the currently selected cell
2. Shift+Enter: Run the currently selected cell + advance to next (if at the
end of the notebook, adds a new cell below)
3. M: Change currently selected cell to Markdown format
4. R: Change currently selected cell to raw format
5. Y: Change currently selected cell to code format
6. A: Add cell above currently selected cell
7. B: Add cell below currently selected cell
8. X: Cut currently selected cell
9. V: Paste currently selected cell
10. Enter: Begin typing in currently selected cell
11. Esc: Stop typing in current cell without executing any code
12. Up arrow / Down arrow: Navigate between cells
13. Ctrl+Shift+Q: Close and shut down current notebook

### Some python methods
- **.find()** searches a string for a specified set of characters (could
be a word or phrase), and returns its location using the index
 Example
`order_status = "order #P080624 is ready for pickup"
print(order_status.find("is ready"))`
output: 20
- **.rfind** works like find, but searches from right to left, still returns starting location of the positive index
*output*: -1 means (substring not found)
- **The format()** function formats the specified value(s)
First parameter is the value to format, second parameter is format type
When using the
 - **.format()** method, a placeholder value can be
inserted into the string using curly brackets {}, then defined in
parentheses when the format method is called
◊ Example:
txt1 = "The ribeye is {price:.2f} dollars, yikes"
print(txt1.format(price = 49))
- **.replace()** method replaces a string with another string:
◊ Example:
a = "Hello, World!"
print(a.replace("H", "J").replace(", ", ""))
output: JelloWorld!
- **The range()** function returns a sequence of numbers
◊ Syntax:
range(start, end , step)
∗ start is optional, defaults to 0
∗ step is optional, defaults to 1




### Math module
 - **math.fsum()** returns the sum of all items in an iterable (tuples, arrays,
lists, etc.)
 - **math.sqrt()** returns the square root of a number
 - **math.isqrt()** rounds the square root of a number down to the nearest
integer
 - **math.fabs()** returns the absolute value of a number
 - **math.fmod()** returns the remainder of x/y (like % operator)
 - **math.floor()** method rounds a number down to the nearest integer;
 - **math.ceil()** rounds a number upwards to the nearest integer; and
 - **math.trunc()** returns the truncated integer of a number
 - **math.pi** constant returns the value of pi (3.14159...)

 

### Statistics module
- **statistics.mean()** calculates the mean (average) of the given data
- **statistics.median()** calculates the median (middle value) of the given data
- **statistics.mode()** calculates the mode (most frequently occurring value)
of the given numeric or nominal data
- **statistics.stdev()** calculates the standard deviation from a sample of data
- **statistics.variance()** calculates the variance from a sample of data

### Random module
- **random.randint()** returns a random integer from a given range
- **random.random()** returns a random float number between 0 and 1
- **random.choice()** returns a random element from a given sequence;
- **random.choices()** returns a list with a random selection from a given
sequence
- **random.sample()** returns a sample of a given sequence (list, set,
range, etc.)
- **random.shuffle()** takes a sequence and returns it in a random order

### Datetime module
- **datetime.datetime()** allowsto write in year-month-day format
- **.strftime()** method is used to format date objects into strings
