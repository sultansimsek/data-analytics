# Exercise 2.A Using built-in functions
## Lab 2. Parsing strings – part codes

```python
def product_info(part_code):
    
  def get_supplier_code():
    return part_code.split(":")[0]

  def get_product_num():
    return part_code.split(":")[1].split("-")[0]

  def get_size():
    return part_code.split("-")[1]

  product = [
        {
            "supplier":  get_supplier_code(),
            "model": get_product_num(),
            "size": get_size()
        }
    ]
  return product
```

```python
result = product_info("ACME:123-L")

print(result)

result = product_info("DI:12345-M")

print(result)

result = product_info("ACE:1-12")

print(result)

```
