Working with defining custom functions.

```python
def favorite_things(name, movie, music):
   return (f"My name is {name}\n"
            f"My favorite movie is {movie}\n"
            f"My favorite musician/band is {music}")

print(favorite_things("Sultan", "Eternity and a Day", "Billie Eilish"))
```

```python
def why_im_here(purpose):
    return purpose

print(why_im_here("I joined Year Up United to build data skills and launch a career in tech."))
```

```python
def favorite_place(place, why):
    return f"One of my favorite places is {place} because {why}."

print(favorite_place("Chicago", "the food scene and architecture are incredible"))


print(favorite_place("the beach", "it helps me relax and clear my mind"))
```
