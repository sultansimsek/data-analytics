# Tax Functions Lab — Week 6

```python
def get_soc_sec_tax(gross):
    rate = 0.062
    return gross * rate
```

```python
def get_medicare_tax(gross):
    rate = 0.0145
    return gross * rate
```

```python
def get_federal_tax(gross, code):
    if code == 0:
        rate = 0.23
    elif code == 1:
        rate = 0.21
    elif code == 2:
        rate = 0.195
    elif code == 3:
        rate = 0.185
    elif code == 4:
        rate = 0.18
    else:
        rate = 0.18 - ((code - 4) * 0.005)
    return gross * rate
```

```python
gross1 = 750
code1 = 0
print(f"--- Person 1 (Gross: ${gross1}, Code: {code1}) ---")
print(f"Social Security Tax: ${get_soc_sec_tax(gross1):.2f}")
print(f"Medicare Tax:        ${get_medicare_tax(gross1):.2f}")
print(f"Federal Tax:         ${get_federal_tax(gross1, code1):.2f}")
```
