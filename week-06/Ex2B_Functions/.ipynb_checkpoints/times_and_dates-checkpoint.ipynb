```python
import datetime
```

### Lab 2 of Exercise 2.B
Working with datetime functions.

```python
today = datetime.datetime.now()
```

```python
print(today)
```

```python
today.strftime("%m %d,%Y")
```

```python
print(f"Today is Friday, {today.strftime('%m %d,%Y')}.")
```

### My birthday

```python
birthday = datetime.date(2001, 0o2, 26)
```

```python

print(f"My birthday is {birthday.strftime('%m %d,%Y')}")
```

### Next 90 Days

```python
ninety_d = today + datetime.timedelta(days=90)
```

```python
print(f"90 days from today is {ninety_d.strftime('%m %d, %y')}.")
```

### Time for dinner

```python
dinner_time = datetime.time(7)
```

```python
print(f"Let’s meet for dinner at {dinner_time.strftime('%H:%M PM')}.") 
```
